```python
#cleaing data missing values

df=pd.read_csv("raw_data.csv")
df.isnull()

df.isnull().sum()
df.dropna()
df.dropna(axis=1)
df.fillna(0)
age_mean=df["age"].mean()

cleaned_data=df.copy()
age_mean=cleaned_data["age"].mean()
cleaned_data["age"]=cleaned_data["age"].fillna(age_mean)
# cleaned_data

df.ffill()
df.bfill()
```

```python
#handling duplicates

df[["country","gender"]].duplicated()
df2=df.copy()
df.drop_duplicates(inplace=True)

```
