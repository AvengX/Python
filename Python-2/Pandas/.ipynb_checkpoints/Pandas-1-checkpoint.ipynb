```python
import pandas as pd
```

```python
info={
    "Name":["Rahul","Harsh","Navya"],
    "CGPA":[9.5,7.5,8.5]
}
df=pd.DataFrame(info)
print(df)
```

```python
s=pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#index
print(s[0])
print(s[1])
```

```python
s=pd.Series([21,42,33,14,75,"abc"],index=["Ayush","Nikhil","Vignesh","Aryan","Siddhant",25])
print(s)

print(s["Ayush"])
print(s.index)
```

```python
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([6,8,7,9,10])
print(s1+s2)

s1[0]=100
changed_s1=s1.drop(0)
print(s1)
print(changed_s1)
```

```python
#dataframes

info={
    "Name":["Ayush","Dubey","Vigoo"],
    "Age":[23,24,25],
    "GPA":[9.0,7.9,9.5]
}
df=pd.DataFrame(info)
print(df,type(df))

print(df.index)
print(df.columns)
```

```python
df=pd.DataFrame([["Ayush",23],["Vignesh",22],["Siddhant",24]],columns=["Name","Age"])
print(df)
```

```python
#creating dataframe using numpy
import numpy as np
np_arr=np.array([[1,2,3],[4,5,6]])
df=pd.DataFrame(np_arr,columns=["A","B","C"])
print(df)
```
